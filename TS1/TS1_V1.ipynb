## Simulación de la Función Transferencia normalizada

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue Apr 11 11:41:36 2023

@author: gonzalo
"""

from scipy.signal import TransferFunction

## Libreria TC2
from pytc2.sistemas_lineales import pzmap, bodePlot

R1n = 2 
conjunto_R2n= [1, 2, 4]

for i in range(len(conjunto_R2n)):
    
    my_tf = TransferFunction([1, -(conjunto_R2n[i])/R1n], [1, 1])

    bodePlot(my_tf, fig_id=1, filter_description = 'R2n/R1n={:3.3f}'.format(conjunto_R2n[i]/R1n) )
    
    pzmap(my_tf, fig_id=2, filter_description = 'R2n/R1n={:3.3f}'.format(conjunto_R2n[i]/R1n) )
```

# Simulación de la Transferencia sin normalizar

Para la simulación de la función de transferencia obtenida en el primer punto, se utilizarán los siguientes valores de componentes:

\begin{equation}

pregunta: ¿Cómo soluciono el problema con las escalas?

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Wed Apr 12 22:22:23 2023

@author: gonzalo
"""
import matplotlib.pyplot as plp
from scipy.signal import TransferFunction
## Libreria TC2
from pytc2.sistemas_lineales import pzmap, bodePlot

# Asigno los valores de componentes 
R1 = 1e3
R2 = 1e3
R3 = 1e3
C1 = 1e-6

#Armo mi Función Transferencia 
w0 = 1/(R3 * C1)
tf = TransferFunction([1, -(R2/R1)*w0], [1, w0])

bodePlot(tf, fig_id=1)    
pzmap(tf, fig_id=2)

#Falta correccion de escalas
```
